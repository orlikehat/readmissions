# Admission

```python
df_adm["LABEL_HOSP"]=np.where(df_adm["LABEL_HOSP"]==1,"readmission","no_readmission")
df_adm["LABEL_JUST_ER"]=np.where(df_adm["LABEL_JUST_ER"]==1,"readmission","no_readmission")

df_1=df_adm[df_adm["LABEL_HOSP"]=="readmission"]
df_0=df_adm[df_adm["LABEL_HOSP"]=="no_readmission"]

df_1_ER=df_adm[df_adm["LABEL_JUST_ER"]=="readmission"]
df_0_ER=df_adm[df_adm["LABEL_JUST_ER"]=="no_readmission"]
```

```python
 def diag_top_20(df,name):
    
    sum_table = df.values.sum(axis=0)   
    sum_table = pd.Series(sum_table, index=df.columns)
    sum_table.to_frame()

    sum_table=sum_table.drop(index='CaseNum')
    sum_table=sum_table.drop(index='NGeneral symptoms and signs')
    sum_table=sum_table.drop(index='NUNSPECIFIED')
    sum_table=sum_table.drop(index='LABEL_HOSP')
    sum_table=sum_table.drop(index='LABEL_JUST_ER')
    
    sum_table=sum_table.astype(int)


    sum_table=sum_table.sort_values(ascending=False)
    sum_table_top=sum_table.iloc[1:20]
        
    sum_table_top.to_csv(output_path+name+"_top20.csv")    
    
```

```python
df_sum_all_adm=diag_top_20(df_adm,"admission_all")
df_sum_all_adm=diag_top_20(df_1,"HOSP_read")
df_sum_all_adm=diag_top_20(df_0,"HOSP_no_read")
df_sum_all_adm=diag_top_20(df_1_ER,"ER_read")
df_sum_all_adm=diag_top_20(df_0_ER,"ER_no_read")

```
